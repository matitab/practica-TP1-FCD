```python
x=[1,2,3,5]
y=[2,3,6,9]

def interpolar(x, y, xo):
    puntos=sorted(zip(x,y)) # con esto le damos el sentido de pares de x,y 
    for i in range(len(puntos)-1): # porque arranca de cero
        if xo>=puntos[i][0] and xo <= puntos[i+1][0]: # vamos a mirar en las X
            y_int= puntos[i][1]+ (xo-puntos[i][0]) * (puntos[i+1][1]-puntos[i][1])/(puntos[i+1][0]-puntos[i][0]) # ecuación de interpolación 
        else:
            y_int='indeterminada por ser imposible interpolar'
    return y_int
```
